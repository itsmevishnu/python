## Binomial distribution
A discrete probability distribution which provide number of successs of _n_ independent experiments.
## Poisson distribution
A discrete probability distribution which give the limit of the binomial distribution.

```python
binom = random.binomial(n=1000, p=0.01, size=1000)
po = random.poisson(lam=10, size=1000)
exp = random.exponential(size=1000)
```

```python
sns.histplot(data=binom, kde=True, color='r')
sns.histplot(data=po, kde=True, color='b')
sns.histplot(data=exp, kde=True, color='g')
```

## Linear regression with Numpy

```python
a = 10
b = 1
x = np.linspace(0,1,100)
```

```python
y = a*x+b+random.randn(len(x))
```

```python
p = np.polyfit(x,y,1)
```

```python
yl = np.polyval(p,x)
yl
```

```python
plt.scatter(x, y, color="r", marker="x")
plt.plot(x,yl, color='b')
```

```python
from functools import reduce
```

```python


def get_sub(input, location):
    try:
        return reduce(dict.get, location, input)
    except TypeError:
        pass

```

```python
json = {
    "name":"Vishnu",
    "age": 30,
    "friends": [{
        "name": 'Abhay',
        'age': 26,
        'dep' : 'Physics'
    },
    {
        "name": 'Semin',
        'age': 29,
        'dep' : 'Physics'
    }
    ]
}
print(get_sub(json, ['friends', 'name']))
```
